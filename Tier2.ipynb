```python
from sqlalchemy import create_engine
import pandas as pd
```

Got this database from the Tier 1 project. Spent way too much time trying to figure out how to get a useable sqlite database from the MySQL database, so just got this one.

```python
engine = create_engine('sqlite:///sqlite_db_pythonsqlite.db')
```

```python
query = """
SELECT
	strftime('%m', b.starttime) AS month,
	f.name AS facility,
	b.bookid
FROM Bookings as b
LEFT JOIN Facilities AS f
	ON b.facid = f.facid
LEFT JOIN Members AS m
	ON b.memid = m.memid
WHERE b.memid <> 0
GROUP BY month, f.name;
"""
```

```python
df = pd.read_sql_query(query, engine)
df
```
